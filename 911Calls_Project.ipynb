```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('911.csv')
df.info
```

```python
df.head()
```

```python
df['zip'].value_counts().head(5)
```

```python
df['twp'].value_counts().head(5)
```

```python
# returns the number of unique tit
df['title'].nunique()
```

```python
x = df['title'].iloc[0]
x
```

```python
x.split(':')[0]
```

```python
df['Reasons'] = df['title'].apply(lambda title : title.split(':')[0])
df['Reasons']
```

```python
df['Reasons'].value_counts()
```

```python
sns.countplot(x='Reasons', data = df, palette = 'viridis')
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
type(df['timeStamp'].iloc[0])
```

```python
time = df['timeStamp'].iloc[0]
time.hour
```

```python
time.dayofweek
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time : time.hour)
df['Hour']
```

```python
df['month'] = df['timeStamp'].apply(lambda time : time.month)
df['month']
```

```python
df['Day of Week'] = df['timeStamp'].apply(lambda time : time.dayofweek)
df['Day of Week']
```

```python
df.head()
```

```python
dmap = {0:'Mon',1:'Tues',2:'Wed',3:'Thrus',4:'Fri',5:'Sat',6:'Sun'}
df['Day of Week'] = df['Day of Week'].map(dmap)
df.head()
```

```python
sns.countplot(x = 'Day of Week', data = df,hue = 'Reasons', palette='viridis')
#to relocate the legend
plt.legend(bbox_to_anchor=(1.05,1),loc=2, borderaxespad=0.)
```

```python
sns.countplot(x = 'month', data = df,hue = 'Reasons', palette='viridis')
#to relocate the legend
plt.legend(bbox_to_anchor=(1.05,1),loc=2, borderaxespad=0.)
```

```python
byMonth = df.groupby('month').count()
byMonth.head()
```

```python
byMonth['lat'].plot()
```

```python
sns.countplot(x = 'month', data = df, palette='viridis')
#to relocate the legend
plt.legend(bbox_to_anchor=(1.05,1),loc=2, borderaxespad=0.)
```

```python
sns.lmplot(x='month',y='twp',data=byMonth.reset_index())
```

```python
t = df['timeStamp'].iloc[0]
t
```

```python
df['Date'] = df['timeStamp'].apply(lambda t:t.date())
df.head()
```

```python
df.groupby('Date').count()['lat'].plot()
plt.tight_layout()
```

```python
df[df['Reasons']=='Traffic'].groupby('Date').count()['lat'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reasons']=='Fire'].groupby('Date').count()['lat'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
df[df['Reasons']=='EMS'].groupby('Date').count()['lat'].plot()
plt.title('EMS')
plt.tight_layout()
```

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reasons'].unstack()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='viridis')
```

```python
plt.figure(figsize=(12,6))
sns.clustermap(dayHour,cmap='viridis')
```

```python
dayMonth = df.groupby(by=['Day of Week','month']).count()['Reasons'].unstack()
dayMonth.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth,cmap='viridis')
```

```python
plt.figure(figsize=(12,6))
sns.clustermap(dayMonth,cmap='viridis')
```
